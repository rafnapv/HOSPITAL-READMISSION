TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x=a.drop('readmission_rate',axis='columns')
x
```

```python
y=a.readmission_rate
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=60)
xtrain,xtest,ytrain,ytest
```

```python
model1.predict(xtest)
```

```python
model1.score(xtest,ytest)
```

VISUALIZATION

Time series: Number of patients over time


```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(14,5))
sns.lineplot(data=a, x="date", y="num_patients_admitted", color="blue")
plt.title("Number of Patients Admitted Over Time", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Patients Admitted")
plt.show()
```

Time series: Average length of stay

```python
plt.figure(figsize=(14,5))
sns.lineplot(data=a, x="date", y="avg_length_of_stay", color="green")
plt.title("Average Length of Stay Over Time", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Average Length of Stay (days)")
plt.show()
```

Correlation heatmap

```python
plt.figure(figsize=(8,6))
corr = a.drop(columns=['date']).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Hospital Metrics", fontsize=14)
plt.show()
```

Pairplot to see relationships between variables

```python
sns.pairplot(a.drop(columns=['date']), diag_kind="kde")
plt.suptitle("Relationships Between Variables", y=1.02)
plt.show()
```

Distribution of readmission rate

```python
plt.figure(figsize=(8,5))
sns.histplot(a["readmission_rate"], bins=30, kde=True, color="purple")
plt.title("Distribution of Readmission Rate", fontsize=14)
plt.xlabel("Readmission Rate")
plt.ylabel("Frequency")
plt.show()
```
